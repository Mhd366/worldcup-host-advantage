3. Data Cleaning

```python
df = pd.read_csv('../data/raw/worldcup_matches.csv')

# --- 3.1 Keep only analysis-relevant columns ---
keep_cols = ['Year', 'Host', 'Date', 'Round', 'Venue', 'Attendance',
             'home_team', 'away_team', 'home_score', 'away_score',
             'home_penalty', 'away_penalty',
             'home_red_card', 'away_red_card']
df = df[keep_cols].copy()

```

```python
# --- 3.2 Rename to consistent snake_case ---
df.columns = df.columns.str.lower()

```

```python
# --- 3.3 Convert data types ---
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['attendance'] = pd.to_numeric(df['attendance'], errors='coerce')

```

```python
# --- 3.4 Handle missing values (with justification!) ---
# penalties: NaN means "no shootout" -> flag it, then drop raw cols
df['went_to_shootout'] = df['home_penalty'].notna().astype(int)
# red cards: NaN means "no red card" -> count of names, 0 if NaN
df['home_red_cards'] = df['home_red_card'].fillna('').apply(
    lambda x: 0 if x == '' else len(str(x).split('|')))
df['away_red_cards'] = df['away_red_card'].fillna('').apply(
    lambda x: 0 if x == '' else len(str(x).split('|')))
df = df.drop(columns=['home_penalty', 'away_penalty',
                      'home_red_card', 'away_red_card'])

```

```python
# --- 3.5 Harmonize historical country names ---
name_map = {'West Germany': 'Germany', 'FR Yugoslavia': 'Serbia',
            'Yugoslavia': 'Serbia', 'Soviet Union': 'Russia',
            'Czechoslovakia': 'Czech Republic', 'Dutch East Indies': 'Indonesia'}
for col in ['home_team', 'away_team', 'host']:
    df[col] = df[col].replace(name_map)

```

```python
# --- 3.6 Feature engineering (قلب المشروع) ---
df['total_goals'] = df['home_score'] + df['away_score']
df['result'] = np.select(
    [df['home_score'] > df['away_score'],
     df['home_score'] < df['away_score']],
    ['home_win', 'away_win'], default='draw')
# هل أحد الطرفين هو المستضيف؟
df['host_playing'] = ((df['home_team'] == df['host']) |
                      (df['away_team'] == df['host'])).astype(int)
df['host_won'] = np.where(df['host_playing'] == 1,
    ((df['home_team'] == df['host']) & (df['result'] == 'home_win')) |
    ((df['away_team'] == df['host']) & (df['result'] == 'away_win')), np.nan)

```

```python
# --- 3.7 Sanity checks ---
print(df.shape)
print(df.isnull().sum())
print(df['host_playing'].sum(), "مباراة لعبها المستضيف")

```

```python
# --- 3.8 Save ---
df.to_csv('../data/clean/clean_data.csv', index=False)
```

```python
# ============ Visual Identity (Saudi 2034 theme) ============

COLORS = {'green': '#006C35', 'gold': '#C9A227',
          'gray': '#8B949E', 'dark': '#0E1117'}
sns.set_theme(style='whitegrid', palette=[COLORS['green'], COLORS['gold'], COLORS['gray']])
plt.rcParams.update({'figure.figsize': (10, 5), 'axes.titlesize': 14,
                     'axes.titleweight': 'bold', 'figure.dpi': 110})
```

```python
# --- Viz 1: Goals per match (Histogram) ---
fig, ax = plt.subplots()
sns.histplot(df['total_goals'], bins=range(0, 13), color=COLORS['green'], ax=ax)
ax.set_title('Distribution of Total Goals per Match (1930–2022)')
ax.set_xlabel('Total goals'); ax.set_ylabel('Matches')
plt.savefig('../assets/viz1_goals_hist.png', bbox_inches='tight')
plt.show()
```

Most World Cup matches produce **2–4 goals**, with the distribution right-skewed: high-scoring games (8+) are rare and mostly from early eras. This defines what a "typical" match looks like before we isolate the host effect.

```python
# --- Viz 2: Matches by round (Countplot) ---
fig, ax = plt.subplots()
order = df['round'].value_counts().index
sns.countplot(data=df, y='round', order=order, color=COLORS['green'], ax=ax)
ax.set_title('Matches by Tournament Stage (1930–2022)')
ax.set_xlabel('Matches'); ax.set_ylabel('')
plt.savefig('../assets/viz2_rounds_count.png', bbox_inches='tight'); plt.show()
```

Clear first evidence of host advantage: median attendance in host matches (~70k) far exceeds other matches (~40k), and the entire host IQR sits above the non-host box. Outliers above 150k are historic finals (e.g., Maracanã 1950). Bigger crowds are the mechanism through which home support could translate into performance.

```python
# --- Viz 3: Attendance — host matches vs others (Boxplot) ---
fig, ax = plt.subplots()
sns.boxplot(data=df, x='host_playing', y='attendance',
            palette=[COLORS['gray'], COLORS['green']], ax=ax)
ax.set_xticklabels(['Host not playing', 'Host playing'])
ax.set_title('Attendance: Host Matches Draw Bigger Crowds')
plt.savefig('../assets/viz3_attendance_box.png', bbox_inches='tight')
plt.show()
```

```python
# --- Viz 4: Attendance over time (Scatter) ---
fig, ax = plt.subplots()
sns.scatterplot(data=df, x='year', y='attendance', hue='host_playing',
                palette=[COLORS['gray'], COLORS['green']], alpha=0.6, ax=ax)
ax.set_title('World Cup Attendance Growth (1930–2022)')
ax.set_xlabel('Year'); ax.set_ylabel('Attendance')
ax.legend(labels=['Host not playing', 'Host playing'])
plt.savefig('../assets/viz4_attendance_scatter.png', bbox_inches='tight'); plt.show()
```

```python
# --- Viz 5: THE GOLDEN CHART — Host win rate vs baseline (Bar) ---
host_matches = df[df['host_playing'] == 1]
host_win_rate = host_matches['host_won'].mean()
baseline = (df.loc[df['host_playing'] == 0, 'result'] != 'draw').mean() / 2  # متوسط فوز طرف واحد

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(['Average team', 'Host nation'], [baseline, host_win_rate],
              color=[COLORS['gray'], COLORS['green']])
ax.bar_label(bars, fmt='%.0%%' if hasattr(ax, '') else lambda v: f'{v:.0%}')
ax.set_title('Host Nations Win Far More Often Than Average')
ax.set_ylabel('Win rate'); ax.set_ylim(0, 1)
plt.savefig('../assets/viz5_host_winrate.png', bbox_inches='tight'); plt.show()
print(f"Host win rate: {host_win_rate:.1%} | Baseline: {baseline:.1%}")
```

**The headline finding:** host nations win **62.7%** of their matches vs. a **38.6%** baseline — a +24pp gap. Part of this reflects that some hosts are strong teams anyway, but the size of the effect (and cases like South Korea 2002 or Russia 2018 overperforming their history) points to a genuine host advantage. This motivates including `host_playing` as a feature in our predictive model.

```python
# --- Viz 6: Correlation heatmap ---
num_cols = ['year', 'attendance', 'home_score', 'away_score',
            'total_goals', 'host_playing', 'went_to_shootout']
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f',
            cmap=sns.light_palette(COLORS['green'], as_cmap=True), ax=ax)
ax.set_title('Correlation Matrix of Numeric Features')
plt.savefig('../assets/viz6_corr_heatmap.png', bbox_inches='tight'); plt.show()
```
